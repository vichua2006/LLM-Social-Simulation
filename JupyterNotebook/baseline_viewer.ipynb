```python
#all
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
filepath = 'D:\StudyLife\Github\LLM-Social-Simulation\CSV\CleanCSV.csv'
df = pd.read_csv(filepath)
df = df[df.iloc[:, 0]!='Average'].iloc[:, 1:][:4]
#drop the first 5 columns
df = df.iloc[:, 11:]
df =  df.apply(pd.to_numeric, errors='ignore')
df
```

```python
df["Total Trade Count"] = df["Trade Count Before"] + df["Trade Count After"]
df["Total Trade Accepted Count"] = df["Trade Accepted Count Before"] + df["Trade Accepted Count After"]
df["Total Rob Count"] = df["Rob Count Before"] + df["Rob Count After"]
df["Total Rob Rebelled Count"] = df["Rob Rebelled Count Before"] + df["Rob Rebelled Count After"]
df["Total Farm Count"] = df["Farm Count Before"] + df["Farm Count After"]
df["Total Count Before"] = df["Trade Count Before"] + df["Rob Count Before"] + df["Farm Count Before"]
df["Total Count After"] = df["Trade Count After"] + df["Rob Count After"] + df["Farm Count After"]
df["Total Count"] = df["Total Count Before"] + df["Total Count After"]
df["Total Trade Rate"] = df["Total Trade Count"] / df["Total Count"]
df["Total Trade Accepted Rate"] = df["Total Trade Accepted Count"] / df["Total Count"]
df["Total Rob Rate"] = df["Total Rob Count"] / df["Total Count"]
df["Total Rob Rebelled Rate"] = df["Total Rob Rebelled Count"] / df["Total Count"]
df["Total Farm Rate"] = df["Total Farm Count"] / df["Total Count"]
df
```

```python
if 'Average' not in df.index:
    df.loc['Average'] = df.mean()
convergence_time = df.loc['Average', 'Convergence Time (days)']
df
```

```python
df.to_csv('baseline.csv', index=False)
```
