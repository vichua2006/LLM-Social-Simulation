```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the relationships
relationships = {
    'A': [('B', 'rob')],
    'B': [('C', 'trade with')],
    'C': [('A', 'talk with')]
}

# Create a directed graph
G_colored = nx.DiGraph()

# Define colors for each type of relationship
edge_colors = {
    'rob': 'red',
    'trade with': 'green',
    'talk with': 'blue'
}

# Add nodes and edges with labels and colors
for person, interactions in relationships.items():
    for target, action in interactions:
        G_colored.add_edge(person, target, label=action, color=edge_colors[action])

# Position the nodes using the spring layout algorithm
pos = nx.spring_layout(G_colored)

# Extract colors from the graph
colors = [edata['color'] for f, t, edata in G_colored.edges(data=True)]

# Draw the graph with specified node positions and colored edges
nx.draw(G_colored, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=10, font_weight='bold', edge_color=colors)

# Draw edge labels
edge_labels_colored = nx.get_edge_attributes(G_colored, 'label')
nx.draw_networkx_edge_labels(G_colored, pos, edge_labels=edge_labels_colored, font_color='black')

# Show the graph
plt.title("Relationship Graph with Colored Edges")
plt.show()
```
